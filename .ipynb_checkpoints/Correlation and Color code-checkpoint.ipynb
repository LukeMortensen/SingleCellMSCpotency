```python

def bar_plots(temp):
    bardf=temp.sort_values(temp.columns[0], ascending = False).T
    ax=sns.barplot(data=bardf,orient="h", ci=None)
    title='Correlations of '+temp.columns[0]+'[%]'
    ax.set_title(title)
    fig_name="barplot_"+temp.columns[0]+".png"
    ax.figure.savefig(fig_name,dpi=300,bbox_inches="tight")
    
df_loc="C:/Users/priya/OneDrive - University of Georgia/Projects/MALDI/MALDI_Gtech/MALDI_7thtrial_Singlecell/Coregistartion/Cellprofiler_S1W2_IFN/"
EXT="*.xlsx"
allFiles=[]
#df_mad = pd.DataFrame()
df_mean = pd.DataFrame()
df_integrated = pd.DataFrame()
for path, subdir, files in os.walk(df_loc):
    for file in glob(os.path.join(path,EXT)):
        allFiles.append(file)

for i in allFiles:
    df=pd.read_excel(i)
    print(i.split("/")[-1],df.shape)
    dfmat=df.drop(['Group', ], axis=1)
    correlated_features = set()
    correlation_matrix = dfmat.astype(float).corr()
    for i in range(len(correlation_matrix.columns)):
        for j in range(i):
            if abs(correlation_matrix.iloc[i, j]) > 0.80: # play around with r
                colname = correlation_matrix.columns[i]
                correlated_features.add(colname)
    correlation_matrix[np.isnan(correlation_matrix)] = 0
    sub_corr=correlation_matrix[['Area','BoundingBoxArea','Compactness','Eccentricity','Extent','FormFactor','MajorAxisLength','MaxFeretDiameter','MaximumRadius','MeanRadius',
                'MedianRadius','MinFeretDiameter','MinorAxisLength','Perimeter','Solidity']][16:]
    bar_plots(sub_corr[1:].T)
    #bar_plots(sub_corr[1].T)
    print(sub_corr[1:])
    df_mean=df_mean.append(sub_corr[1:]) # MAD 1, Mean 2
    #df_mad=df_mad.append(sub_corr.iloc[:1]) # Only MAD 1
    #df_mean=df_mean.append(sub_corr.iloc[1]) # Only Mean 2
    #df_integrated=df_integrated.append(sub_corr[:1])
    #allData=allData.append(f1, ignore_index = True)
#print(allData.shape)
```
